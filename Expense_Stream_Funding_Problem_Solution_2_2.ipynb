```
#Problem part B

import numpy as np
import cvxpy as cp
n = 36

e = np.matrix([
  0.000000,
  1.000000,
  2.000000,
  3.000000,
  4.000000,
  5.000000,
  6.000000,
  7.000000,
  8.000000,
  16.000000,
  16.600000,
  17.200000,
  17.800000,
  18.400000,
  19.000000,
  19.600000,
  20.200000,
  18.000000,
  17.000000,
  16.000000,
  15.000000,
  14.000000,
  13.000000,
  12.000000,
  11.000000,
  10.000000,
  9.000000,
  8.000000,
  7.000000,
  6.000000,
  5.000000,
  4.000000,
  3.000000,
  2.000000,
  1.000000,
  0.000000,
]).T
 
rho = 0.002500

#xb = cp.Variable(shape = (m,1))
w = cp.Variable(shape = (n,1))
b = cp.Variable(shape = (n,1))
 
#for i in range(n-1):
#  b[i+1]==((1+rho)*b[i])-w[i] 

objective = cp.Minimize(b[1])

#Constraints
constraints = [b>=0, w>=e]
constraints += [b[i+1]==((1+rho)*b[i])-w[i] for i in range(n-1)]
    #{b[i+1]==((1+rho)*b[i])-w[i] for i in range(n-1)}
 
prob = cp.Problem(objective, constraints)
prob.solve()
 
#Answers
#print("P:", P.shape)
#print("e:", e.shape)
#print("one:", one.shape)
#print("x:", x.shape)
#print("w:", w.shape)
#print("b:", b.shape)
print("Minimum initial investment required when no investments are purchased:", b[1].value)
#print("Investments purchased initially (x):", x.value)
print("Bank balances per year (b):", b.value)
print("Withdrawals per year: (w)", w.value)
```
